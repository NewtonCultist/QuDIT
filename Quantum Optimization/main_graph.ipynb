```python
##experimento control optimizado por qiskit nivel 3

from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, transpile
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_tools import creacion_GHZ, state_3layers

shot = 50000

max_qbits = 0
max_len_qbits = 0
for i in lista_de_qbits:
    if max(i) > max_qbits:
        max_qbits = max(i)
    if max_len_qbits < len(i):
        max_len_qbits = len(i)

q_r = QuantumRegister(max_qbits + 1, "q")
circuitos = list()

for j in lista_de_qbits:
    circuitos.append(state_3layers(len(j) ,j, q_r))
    
service = QiskitRuntimeService(channel= "ibm_quantum", token="f357e84602c6054693aa4b405885636006fd0814819b4b5be7c2839059ae57b029bdf5092d66b9d84d0878d9daa7ea8f74497eeec41a075c54d59924f03ae020")
computador = service.backend(name=backend)
pm = generate_preset_pass_manager(optimization_level=3, backend=computador)
transpiled = pm.run(circuitos)
job = computador.run(transpiled, shots=shot)
```

```python
job_0 = "cwk8m899r49g00899tp0"
job_1 = "cwk8md140e00008973e0"
job_2 = "cwk8mmj9r49g00899tqg"
job_3 = "cwk939nmptp00085z88g"

import numpy as np

def hellingers(job_id, shots):
    retrieved_job = service.job(job_id)
    results = retrieved_job.result()
    
    prob_experimental = list()
    prob_teorica = list()
    
    for i in range(N_Q - n_q):
        prob_t = list()
        prob_e = list()
        for j in results.results[i].data.counts:
            count = results.results[i].data.counts[j]
            prob_e.append(float(count/shots))
            prob_t.append(float(1/len(results.results[i].data.counts)))
            
        prob_experimental.append(prob_e)
        prob_teorica.append(prob_t)
        
    #print(len(prob_experimental))
    #print(len(prob_teorica))

    distances = list()
    for i in range(len(prob_experimental)):
        distance = np.sqrt(0.5 * np.sum((np.sqrt(np.array(prob_experimental[i])) - np.sqrt(np.array(prob_teorica[i]))) ** 2))
        distances.append(distance)
    
    return distances

print(hellingers(job_0, 50000))
print(hellingers(job_1, 50000))
print(hellingers(job_2, 50000))
print(hellingers(job_3, 50000))

H = [hellingers(job_0, 50000), hellingers(job_1, 50000), hellingers(job_2, 50000), hellingers(job_3, 50000)]
    
```

```python
"""Esta parte del codigo se encarga de graficar las Hellingers"""

import matplotlib.pyplot as plt

claves = list(range(N_Q - n_q))
    
xlabel = list(range(n_q, N_Q))
valores1 = [H[0][clave] for clave in claves]
valores2 = [H[1][clave] for clave in claves]
valores3 = [H[2][clave] for clave in claves]
valores4 = [H[3][clave] for clave in claves]

plt.figure(figsize=(10, 5))
#print(valores1)
#print(valores2)
plt.plot(xlabel, valores1, marker='o', label='Optimizado Direccional 2.0')
plt.plot(xlabel, valores2, marker='s', label='Control')
plt.plot(xlabel, valores3, marker='x', label='Optimizado')
plt.plot(xlabel, valores4, marker='p', label='Optimizado Direccional 1.0')

# Aadir ttulos y etiquetas
plt.title('Comparacin de distancias de hellinger')
plt.xlabel('N Qbits')
plt.ylabel('Distancia de Hellinger')
plt.legend()

# Mostrar el grfico
plt.grid(True)
plt.show()
```
